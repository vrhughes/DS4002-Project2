```python
# Importing packages
import pandas as pd
import re
```

```python
# Reading in the data
gdp_df = pd.read_csv('/content/DS4002-Project2/DATA/GDP-Per-Capita-Current-USD.csv')
```

```python
# Reorganizing data so it matches the other datasets and files

# Ensure consistency in GDP dataset column names
gdp_df.rename(columns={'Country Name': 'Country'}, inplace=True)

# Identify the year columns in GDP data
year_cols = [col for col in gdp_df.columns if re.search(r'\b\d{4}\b', col)]

# Reshape GDP data from wide to long format
gdp_df = gdp_df.melt(id_vars=['Country'], value_vars=year_cols, var_name='Year', value_name='GDP_Per_Capita')

# Convert Year to numeric to match other datasets
gdp_df['Year'] = gdp_df['Year'].str.extract(r'(\b\d{4}\b)').astype(int)
```

```python
# Exporting data
gdp_df.to_csv('gdp_df.csv', index=False)
```
